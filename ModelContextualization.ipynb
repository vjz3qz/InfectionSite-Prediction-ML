```python
#skip SRR1056147 and SRR1056133 and SRR1056148
sequences = ['SRR']*139 #skipping SRR#+92, so that is why it is 138 not 139
for i in range(139):
    num=1056149+i
    num=str(num)
    sequences[i]=sequences[i]+num
    
```

```python
for seq in sequences:
    locals()["dictionary" + str(seq)] = {}
    counter = 0
    
    with open(str(seq) + '.tsv', 'r') as reads:
            for line in reads:
                line = line.split()
                gene_len = float(line[1])
                hits = float(line[2])
                if hits != 0:
                    normalized_numb = (hits * 51) / gene_len
                    seq_name = line[0]
                    seq_name = 'PA14_' + seq_name[-5:]
                    locals()["dictionary"+str(seq)].update( {seq_name : normalized_numb} )
                    counter = counter + 1
```
